```python
import ast
import pandas as pd

opa_new_data = []
with open("movie-wfs-opa-minimal-request-time.dat", 'r') as f:
    for line in f:
        opa_new_data.append(ast.literal_eval(line))

opa_old_data = []
with open("movie-wfs-opa-minimal-request-time-old.dat", 'r') as f:
    for line in f:
        opa_old_data.append(ast.literal_eval(line))
```

```python
df_new = pd.DataFrame(opa_new_data, columns=['src', 'dst', 'time_namelookup', 'time_connect', 'time_appconnect', 'time_pretransfer', 'time_redirect', 'time_starttransfer', 'time_total'])
df_old = pd.DataFrame(opa_old_data, columns=['src', 'dst', 'time_namelookup', 'time_connect', 'time_appconnect', 'time_pretransfer', 'time_redirect', 'time_starttransfer', 'time_total'])


float_cols = ['time_namelookup', 'time_connect', 'time_appconnect', 'time_pretransfer', 'time_redirect', 'time_starttransfer', 'time_total']

for col in float_cols:
    df_new[col] = df_new[col].astype(float)
    df_old[col] = df_old[col].astype(float)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
from pylab import rcParams
from scipy.stats import f_oneway
from scipy.stats import ttest_ind

%matplotlib inline
warnings.filterwarnings("ignore")
rcParams['figure.figsize'] = 20,10
rcParams['font.size'] = 30
sns.set()
np.random.seed(8)

alpha = 0.05
```

```python
def plot_distribution(inp):
    plt.figure()
    ax = sns.distplot(inp)
    plt.axvline(np.mean(inp), color="k", linestyle="dashed", linewidth=5)
    _, max_ = plt.ylim()
    plt.text(
        inp.mean() + inp.mean() / 10,
        max_ - max_ / 10,
        "Mean: {:.2f}".format(inp.mean()),
    )
    return plt.figure
```

```python
plt.figure()
ax1 = sns.distplot(df_new["time_total"])
ax2 = sns.distplot(df_old["time_total"])
plt.axvline(np.mean(df_new["time_total"]), color='blue', linestyle='dashed', linewidth=5)
plt.axvline(np.mean(df_old["time_total"]), color='orange', linestyle='dashed', linewidth=5)
```
